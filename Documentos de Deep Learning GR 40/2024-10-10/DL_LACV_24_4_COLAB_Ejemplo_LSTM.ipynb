```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generación de datos de ejemplo para series de tiempo
# Vamos a crear una serie de tiempo simple para ilustrar el uso del LSTM
def create_dataset(series, time_steps=1):
    X, y = [], []
    for i in range(len(series) - time_steps):
        X.append(series[i:(i + time_steps)])
        y.append(series[i + time_steps])
    return np.array(X), np.array(y)

# Parámetros para la serie de tiempo
data = np.sin(np.linspace(0, 100, 1000))  # Ejemplo de una serie de tiempo sinusoidal
time_steps = 10  # Número de pasos de tiempo para predecir el siguiente valor

# Preparación de los datos
X, y = create_dataset(data, time_steps)
X = X.reshape((X.shape[0], X.shape[1], 1))  # Reshape para ser compatible con LSTM

for i in range(6):
  print("Tiempo i: ",i, "\nX_",i, ": \n", X[i], "  y:", y[i])
```

```python


# Construcción del modelo LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(time_steps, 1)))
model.add(Dense(1))

# Compilación del modelo
model.compile(optimizer='adam', loss='mse')

# Entrenamiento del modelo
model.fit(X, y, epochs=200, verbose=1)

# Evaluación del modelo
predicted = model.predict(X)

# Mostrar las primeras 5 predicciones vs. valores reales
for i in range(5):
    print(f'Real: {y[i]}, Predicción: {predicted[i]}')
```


### Explicación del código

1. **Generación de datos:** Creamos una serie de tiempo sinusoidal como ejemplo y la transformamos en pares de entrada y salida usando una ventana de tiempo (`time_steps`).

2. **Preparación de los datos:** Reestructuramos los datos para que sean compatibles con la entrada requerida por la capa LSTM, que espera una forma de `(muestras, pasos de tiempo, características)`.

3. **Construcción del modelo:** Creamos un modelo secuencial con una capa LSTM y una capa densa de salida para la predicción.

4. **Entrenamiento del modelo:** Entrenamos el modelo en los datos de la serie de tiempo.

5. **Predicciones:** Realizamos predicciones y mostramos una comparación entre algunos valores reales y los predichos.
