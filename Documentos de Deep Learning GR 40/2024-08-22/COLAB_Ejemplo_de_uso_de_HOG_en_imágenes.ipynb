```python
import matplotlib.pyplot as plt
from skimage import data, color, feature
from skimage.transform import rescale

# Cargar una imagen de ejemplo
image = color.rgb2gray(data.astronaut())

# Redimensionar la imagen para acelerar el procesamiento
image_rescaled = rescale(image, 0.5, anti_aliasing=True)

# Extraer características HOG
# Remove the multichannel parameter as it is not supported in this version
hog_features, hog_image = feature.hog(image_rescaled,
                                      orientations=9,
                                      pixels_per_cell=(8, 8),
                                      cells_per_block=(2, 2),
                                      visualize=True)

# Mostrar la imagen original y la imagen con las características HOG
fig, ax = plt.subplots(1, 2, figsize=(12, 6),
                       subplot_kw={'xticks': [], 'yticks': []})

ax[0].imshow(image_rescaled, cmap=plt.cm.gray)
ax[0].set_title('Imagen Original')

ax[1].imshow(hog_image, cmap=plt.cm.gray)
ax[1].set_title('Imagen con características HOG')

plt.show()
```

```python
print("HOG features: ", hog_features)
print("Cantidad de atributos HOG: ", hog_features.shape)
print("Imagen original, tamaño: ", image_rescaled.shape)
```
